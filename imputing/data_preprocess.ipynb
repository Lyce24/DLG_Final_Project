```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv('../data/mutation_matrix_with_clinical_data.csv')
# filter out rows with stage == 'NaN'
df = df[pd.notna(df['Stage'])]  # Keeps rows where 'Stage' is NOT NaN
df
```

```python
print("Unique values in MSI:", df['MSI'].unique())
print("Unique values in Stage:", df['Stage'].unique())

# Replace specific MSI values
df['MSI'] = df['MSI'].replace({
    'Indeterminate': np.nan,
    'Do not report': np.nan,
    'Stable': 0,
    'Instable': 1
}).astype(float)  # Convert to numeric type

# Convert 'Stage' column to string and apply transformation
df['Stage'] = df['Stage'].astype(str)  # Ensure values are strings
df['Stage'] = df['Stage'].apply(lambda x: 1 if x.startswith('4') or x.startswith('3') else 0)

# Check unique values
print("Unique values in MSI:", df['MSI'].unique())
print("Unique values in Stage:", df['Stage'].unique())
df
```

```python
# 1. Basic Information
def basic_info(df):
    print("Dataset Shape:", df.shape)
    print("\nColumn Types:\n", df.dtypes.value_counts())
    print("\nMissing Values per Column:\n", df.isnull().sum().sort_values(ascending=False).head(10))

# 2. Summary Statistics
def summary_statistics(df):
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    summary = df[numerical_cols].describe().T
    print("\nSummary Statistics:\n", summary)

# 3. Missing Value Analysis
def missing_values(df):
    missing_data = df.isnull().sum()
    missing_data = missing_data[missing_data > 0].sort_values(ascending=False)
    print("\nMissing Values in Features:\n", missing_data)
    print(f"\nTotal Missing Values: {df.isnull().sum().sum()}")

# 4. Feature Distributions
def feature_distributions(df):
    print("\nMutation Frequency per Gene (Top 10 Mutated Genes):")
    mutation_counts = df.iloc[:, 1:-2].sum().sort_values(ascending=False)
    print(mutation_counts.head(10))

    print("\nMSI Distribution:\n", df["MSI"].value_counts(dropna=False))
    print("\nStage Distribution:\n", df["Stage"].value_counts())

# 5. Correlation Analysis
def correlation_analysis(df):
    correlation_matrix = df.iloc[:, 1:].corr()
    stage_corr = correlation_matrix["Stage"].sort_values(ascending=False)
    msi_corr = correlation_matrix["MSI"].sort_values(ascending=False)

    print("\nTop 10 Genes Correlated with Stage:\n", stage_corr.head(10))
    print("\nTop 10 Genes Correlated with MSI:\n", msi_corr.head(10))

# Run Analysis
basic_info(df)
summary_statistics(df)
missing_values(df)
feature_distributions(df)
correlation_analysis(df)
```

```python
# Save the cleaned DataFrame to a new CSV file
output_file = '../data/mutation_matrix_cleaned.csv'
df.to_csv(output_file, index=False)
print(f"\nCleaned DataFrame saved to {output_file}")
```
