```python
import pandas as pd
import numpy as np

patients_df = pd.read_csv('../data/msk_2024_mutations__matrix.csv')
patient_mapping = pd.read_csv('../data/patients_mapping.csv')

# change the unnamed:0 column to patient_id
patients_df = patients_df.rename(columns={'Unnamed: 0': 'Patient'})
# drop column Overall_Survival_Months and Vital_Status
patients_df = patients_df.drop(columns=['Overall_Survival_Months', 'Vital_Status'])

# add mapping to the patients_df
patients_df = pd.merge(patients_df, patient_mapping, on='Patient')

# save the patients_df
patients_df.to_csv('../data/msk_2024_mutations_final.csv', index=False)
```

```python
from cbio_py import cbio_mod as cb

patients_df['highest_stage_recorded'] = np.nan

unique_patients = patients_df['Patient'].unique()

for patient in unique_patients:
    patient_data = cb.getAllClinicalDataOfPatientInStudy(studyId='msk_chord_2024', patientId=patient)
    for data in patient_data:
        if data['clinicalAttributeId'] == 'STAGE_HIGHEST_RECORDED':
            patients_df.loc[patients_df['Patient'] == patient, 'highest_stage_recorded'] = data['value']
            break

patients_df

# add highest stage recorded to the patients_df
```

```python
print(patients_df['highest_stage_recorded'].unique())

# get how many patients are in each stage
stage_counts = patients_df['highest_stage_recorded'].value_counts()
print(stage_counts)
```
