```python
import pandas as pd
import numpy as np

# save the patients_df
patients_df = pd.read_csv('../data/msk_2024_mutations_final_version.csv')
```

```python
patients_df
```

```python
from cbio_py import cbio_mod as cb

patient_data = cb.getAllClinicalDataOfPatientInStudy(studyId='msk_chord_2024', patientId='P-0000015')
sample_id = cb.getAllSamplesOfPatientInStudy(studyId='msk_chord_2024', patientId='P-0089676', return_type = 'dict')
print(sample_id)
sample_data = cb.getAllClinicalDataOfSampleInStudy(studyId='msk_chord_2024', sampleId= 'P-0089676-T01-IM7')
sample_data
```

```python
patients_df['CURRENT_AGE_DEID'] = np.nan
patients_df['CNS_BRAIN'] = np.nan
patients_df['LIVER'] = np.nan
patients_df['LUNG'] = np.nan
patients_df['CANCER_TYPE'] = np.nan
patients_df['CANCER_TYPE_DETAILED'] = np.nan
patients_df['MSI_TYPE'] = np.nan
patients_df['CLINICAL_SUMMARY'] = np.nan
patients_df['METASTATIC_SITE'] = np.nan
patients_df['TMB_NONSYNONYMOUS'] = np.nan
patients_df['FRACTION_GENOME_ALTERED'] = np.nan

unique_patients = patients_df['Patient'].unique()

for patient in unique_patients:
    patient_data = cb.getAllClinicalDataOfPatientInStudy(studyId='msk_chord_2024', patientId=patient)
    for data in patient_data:
        
        if data['clinicalAttributeId'] == 'CURRENT_AGE_DEID':
            patients_df.loc[patients_df['Patient'] == patient, 'CURRENT_AGE_DEID'] = data['value']
            
        if data['clinicalAttributeId'] == 'CNS_BRAIN':
            patients_df.loc[patients_df['Patient'] == patient, 'CNS_BRAIN'] = data['value']
            
        if data['clinicalAttributeId'] == 'LIVER':
            patients_df.loc[patients_df['Patient'] == patient, 'LIVER'] = data['value']
            
        if data['clinicalAttributeId'] == 'LUNG':
            patients_df.loc[patients_df['Patient'] == patient, 'LUNG'] = data['value']
        
            
    sample_id = cb.getAllSamplesOfPatientInStudy(studyId='msk_chord_2024', patientId=patient, return_type = 'dict')
    if len(sample_id) > 1:
        print(f"Patient {patient} has multiple samples: {sample_id}")
        
    else:
        sample_id = sample_id[0]['sampleId']
        sample_data = cb.getAllClinicalDataOfSampleInStudy(studyId='msk_chord_2024', sampleId=sample_id)

        for sample in sample_data:

            if sample['clinicalAttributeId'] == 'CANCER_TYPE':
                patients_df.loc[patients_df['Patient'] == patient, 'CANCER_TYPE'] = sample['value']
            if sample['clinicalAttributeId'] == 'CANCER_TYPE_DETAILED':
                patients_df.loc[patients_df['Patient'] == patient, 'CANCER_TYPE_DETAILED'] = sample['value']
            if sample['clinicalAttributeId'] == 'MSI_TYPE':
                patients_df.loc[patients_df['Patient'] == patient, 'MSI_TYPE'] = sample['value']
            if sample['clinicalAttributeId'] == 'CLINICAL_SUMMARY':
                patients_df.loc[patients_df['Patient'] == patient, 'CLINICAL_SUMMARY'] = sample['value']
            if sample['clinicalAttributeId'] == 'METASTATIC_SITE':
                patients_df.loc[patients_df['Patient'] == patient, 'METASTATIC_SITE'] = sample['value']
            if sample['clinicalAttributeId'] == 'TMB_NONSYNONYMOUS':
                patients_df.loc[patients_df['Patient'] == patient, 'TMB_NONSYNONYMOUS'] = sample['value']
            if sample['clinicalAttributeId'] == 'FRACTION_GENOME_ALTERED':
                patients_df.loc[patients_df['Patient'] == patient, 'FRACTION_GENOME_ALTERED'] = sample['value']
                
patients_df
```

```python
patients_df.to_csv('../data/msk_2024_mutations_final_v2.csv', index=False)
```

```python
# non-continuous columns
non_gene_columns = ['highest_stage_recorded', 'CURRENT_AGE_DEID', 'CNS_BRAIN', 'LIVER', 'LUNG', 
                   'CANCER_TYPE', 'CANCER_TYPE_DETAILED', 'MSI_TYPE', 
                   'CLINICAL_SUMMARY', 'METASTATIC_SITE', 
                   'TMB_NONSYNONYMOUS', 'FRACTION_GENOME_ALTERED']

for col in non_gene_columns:
    print(f"Unique values in {col}: {patients_df[col].unique()}")
    counts = patients_df[col].value_counts()
    print(f"Counts for {col}:\n{counts}\n")
```
