```python

# --- 4. Gene Mutation Frequency and Correlation Analysis ---
# identify the gene mutation columns with the highest frequency
gene_mutations = df.iloc[:, 1 : -13]
gene_columns = gene_mutations.columns
print("\nGene mutation columns:", gene_columns.tolist())

# Calculate the frequency of mutations for each gene
mutation_frequencies = gene_mutations.sum() / len(df) * 100
print("\nMutation frequencies (%):")
print(mutation_frequencies.sort_values(ascending=False))
# select the top 10 genes with the highest mutation frequencies
top_genes = mutation_frequencies.sort_values(ascending=False).head(10).index
print("\nTop 10 genes with highest mutation frequencies:")
print(top_genes.tolist())

# Visualize the top 10 gene mutation frequencies with a bar chart
plt.figure()
plt.bar(mutation_frequencies[top_genes].index, mutation_frequencies[top_genes].values)
plt.xlabel("Gene")
plt.ylabel("Mutation Frequency (%)")
plt.title("Top 10 Gene Mutation Frequencies")
plt.xticks(rotation=90)
plt.show()

# Calculate correlation between top 10 gene mutations and OS_MONTHS
correlation_matrix = df[top_genes].corrwith(df['OS_MONTHS'])
print("\nCorrelation between top 10 gene mutations and OS_MONTHS:")
print(correlation_matrix.sort_values(ascending=False))
# Visualize the correlation with a heatmap
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix[top_genes].to_frame(), annot=True, cmap='coolwarm', center=0)
plt.title("Correlation between Top 10 Gene Mutations and OS_MONTHS")
plt.show()
```

```python

# --- 5. Survival Analysis ---
# If you have lifelines installed, you can plot a Kaplan-Meier survival curve.
try:
    from lifelines import KaplanMeierFitter
    kmf = KaplanMeierFitter()

    # Convert OS_STATUS to a binary event indicator: 1 if DECEASED, 0 if LIVING.
    # This assumes OS_STATUS contains the strings "DECEASED" and "LIVING".
    df['event'] = df['OS_STATUS']
    T = df['OS_MONTHS']

    kmf.fit(T, event_observed=df['event'])
    plt.figure()
    kmf.plot_survival_function()
    plt.title("Kaplan-Meier Survival Curve")
    plt.xlabel("Time (months)")
    plt.ylabel("Survival Probability")
    plt.show()
except ImportError:
    print("lifelines package is not installed. Skipping survival analysis.")

# non geneomic data survival analysis: OS_MONTHS	OS_STATUS	highest_stage_recorded	CURRENT_AGE_DEID	CNS_BRAIN	LIVER	LUNG	CANCER_TYPE	TMB_NONSYNONYMOUS	FRACTION_GENOME_ALTERED Regional Distant
non_geneomic_data = df[['OS_MONTHS', 'OS_STATUS', 'highest_stage_recorded', 'CURRENT_AGE_DEID', 'CNS_BRAIN', 'LIVER', 'LUNG', 'TMB_NONSYNONYMOUS', 'FRACTION_GENOME_ALTERED', 'Regional', 'Distant']]
non_geneomic_data

# find correlation between non-geneomic data and OS_MONTHS
correlation_non_geneomic = non_geneomic_data.corr()['OS_MONTHS'].sort_values(ascending=False)
print("\nCorrelation between non-geneomic data and OS_MONTHS:")
print(correlation_non_geneomic)
# Visualize the correlation with a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(non_geneomic_data.corr(), annot=True, cmap='coolwarm', center=0)
plt.title("Correlation between Non-Geneomic Data and OS_MONTHS")
plt.show()
```
