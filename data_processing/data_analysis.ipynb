```python
import pandas as pd

# save the patients_df
df = pd.read_csv('../data/msk_2024_mutations_final.csv')

df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Print basic info about the dataset
print("Data shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Missing values per column:\n", df.isnull().sum())

# --- 2. Descriptive Statistics ---
# Summary statistics for numerical columns
print("\nDescriptive statistics:")
print(df.describe())

# Frequency counts for overall survival status
print("\nOS_STATUS value counts:")
print(df['OS_STATUS'].value_counts())
```

```python

# --- 3. Visualizations ---

# Histogram for overall survival in months
plt.figure()
plt.hist(df['OS_MONTHS'], bins=50, edgecolor='black')
plt.xlabel("Overall Survival (Months)")
plt.ylabel("Number of Patients")
plt.title("Distribution of Overall Survival (OS_MONTHS)")
plt.show()

# Bar chart for OS_STATUS counts
status_counts = df['OS_STATUS'].value_counts()
plt.figure()
plt.bar(status_counts.index, status_counts.values)
plt.xlabel("OS_STATUS")
plt.ylabel("Number of Patients")
plt.title("Patient OS Status Distribution")
plt.show()
```

```python

# --- 4. Gene Mutation Frequency and Correlation Analysis ---
# Assume these columns represent key genes (you can modify this list as needed)
gene_columns = ['TP53', 'KRAS', 'APC', 'PIK3CA', 'EGFR', 'KMT2D','ARID1A', 'KMT2C', 'SMAD4', 'ATM']
print("\nTotal mutation counts for selected genes:")
print(df[gene_columns].sum())

# If there are many mutation subtype columns, you might want to identify them by pattern.
# For example, here we assume that mutation subtype columns contain 'chr' in their names.
mutation_subtype_columns = [col for col in df.columns if ('chr' in col) and (col not in ['OS_MONTHS'])]
print(f"\nNumber of mutation subtype columns: {len(mutation_subtype_columns)}")

# Calculate correlation between each selected gene mutation and OS_MONTHS
correlations = {}
for gene in gene_columns:
    correlations[gene] = df[gene].corr(df['OS_MONTHS'])
print("\nCorrelation between selected genes and OS_MONTHS:")
for gene, corr_val in correlations.items():
    print(f"{gene}: {corr_val:.3f}")

# Visualize correlations with a bar chart
plt.figure()
plt.bar(list(correlations.keys()), list(correlations.values()))
plt.xlabel("Gene")
plt.ylabel("Pearson Correlation with OS_MONTHS")
plt.title("Correlation of Gene Mutations with Overall Survival")
plt.show()
```

```python

# --- 5. Survival Analysis ---
# If you have lifelines installed, you can plot a Kaplan-Meier survival curve.
try:
    from lifelines import KaplanMeierFitter
    kmf = KaplanMeierFitter()

    # Convert OS_STATUS to a binary event indicator: 1 if DECEASED, 0 if LIVING.
    # This assumes OS_STATUS contains the strings "DECEASED" and "LIVING".
    df['event'] = df['OS_STATUS'].apply(lambda x: 1 if 'DECEASED' in str(x) else 0)
    T = df['OS_MONTHS']

    kmf.fit(T, event_observed=df['event'])
    plt.figure()
    kmf.plot_survival_function()
    plt.title("Kaplan-Meier Survival Curve")
    plt.xlabel("Time (months)")
    plt.ylabel("Survival Probability")
    plt.show()
except ImportError:
    print("lifelines package is not installed. Skipping survival analysis.")
```

```python

# --- 6. Comparative Analysis between OS groups ---
# Calculate the average mutation count per gene grouped by OS_STATUS
grouped_mutations = df.groupby('OS_STATUS')[gene_columns].mean()
print("\nAverage mutation counts by OS_STATUS:")
print(grouped_mutations)

# Plotting the average mutation counts for each OS_STATUS
grouped_mutations.plot(kind='bar', figsize=(10, 6))
plt.title("Average Mutation Counts by OS_STATUS")
plt.xlabel("OS_STATUS")
plt.ylabel("Average Mutation Count")
plt.xticks(rotation=0)
plt.show()

# Statistical comparison: perform t-tests for each gene between LIVING and DECEASED groups.
from scipy.stats import ttest_ind

print("\nT-test results comparing mutation counts between LIVING and DECEASED patients:")
for gene in gene_columns:
    living = df[df['OS_STATUS'].str.contains("LIVING", case=False)][gene]
    deceased = df[df['OS_STATUS'].str.contains("DECEASED", case=False)][gene]
    t_stat, p_val = ttest_ind(living, deceased, equal_var=False)
    print(f"{gene}: t-statistic = {t_stat:.3f}, p-value = {p_val:.3e}")
```
