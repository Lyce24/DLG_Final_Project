```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/msk_2024_mutations_final_v2.csv')
df
```

## Filter The Gene Mutation Profile

```python
gene_mutation_profile = df.iloc[:, 1 : 3055]
gene_mutation_profile
```

```python
major_genes = gene_mutation_profile.iloc[:, 0:262]
gene_subtypes = gene_mutation_profile.iloc[:, 262:]
```

```python
# count the number of columns with "_" in the name
num_subtypes = gene_subtypes.columns.str.contains('_').sum()
print(f"Number of subtypes: {num_subtypes}")

numeric_cols = gene_subtypes.select_dtypes(include='number').columns
    
# 3. Calculate the threshold based on the number of rows
threshold = 40

# 4. Filter the numeric columns where the sum is at least the threshold
filtered_subtypes = df[numeric_cols].loc[:, df[numeric_cols].sum() >= threshold]
filtered_subtypes

# patients ID column
patient_df = df.iloc[:, 0:1]
# 5. Concatenate the patient ID column with the filtered subtypes (patient ID is the first column)
filtered_subtypes_with_id = pd.concat([patient_df, filtered_subtypes], axis=1)
filtered_subtypes_with_id
```

## Patient Feature Engineering

```python
feature_columns = df.iloc[:, 3055 : ]
feature_columns = pd.concat([patient_df, feature_columns], axis=1)
feature_columns
```

```python
nan_counts = feature_columns.isna().sum()
nan_counts = nan_counts[nan_counts > 0]
print("Number of NaN values in each feature column:")
print(nan_counts)

# drop columns with too many nan values or unknown values
feature_columns = feature_columns.drop(columns=['METASTATIC_SITE', 'MSI_TYPE', 'CANCER_TYPE_DETAILED'], axis=1)
# count how many nan values are in each feature column

# remove rows with patient ID in P-0006398, P-0020288, P-0017062, P-0045302, P-0053240, P-0075785, P-0077495, P-0004343, they have multiple samples
patients_to_remove = ['P-0006398', 'P-0020288', 'P-0017062', 'P-0045302', 'P-0053240', 'P-0075785', 'P-0077495', 'P-0004343']
feature_columns = feature_columns[~feature_columns.iloc[:, 0].isin(patients_to_remove)]

nan_counts = feature_columns.isna().sum()
nan_counts = nan_counts[nan_counts > 0]
print("Number of NaN values in each feature column after deletion:")
print(nan_counts)
```

```python
# non-continuous columns
for col in feature_columns.columns[1:]:  # Skip the first column (patient ID)
    print(f"Column '{col}' is non-continuous with unique values: {feature_columns[col].unique()}")
    counts = feature_columns[col].value_counts()
    print(f"Counts for {col}:\n{counts}\n")
```

```python
# remove nan values from the feature columns
feature_columns = feature_columns.dropna()
# check for any remaining nan values
print(f'Number of patients left after removing NaN values: {feature_columns.shape[0]}')
```

```python
# clean up the feature table
feature_columns = feature_columns[feature_columns['highest_stage_recorded'] != 'Unknown']
feature_columns = feature_columns[feature_columns['CNS_BRAIN'] != 'Unknown']
feature_columns = feature_columns[feature_columns['LIVER'] != 'Unknown']
feature_columns = feature_columns[feature_columns['LUNG'] != 'Unknown']
feature_columns = feature_columns[feature_columns['CLINICAL_SUMMARY'] != 'Unknown/Unstaged']
feature_columns = feature_columns[feature_columns['CLINICAL_SUMMARY'] != 'Unstaged Unknown']
feature_columns['OS_STATUS'] = np.where(feature_columns['OS_STATUS'] == '1:DECEASED', 1, 0)
feature_columns['highest_stage_recorded'] = np.where(feature_columns['highest_stage_recorded'] == 'Stage 4', 1, 0)
feature_columns['CNS_BRAIN'] = np.where(feature_columns['CNS_BRAIN'] == 'Yes', 1, 0)
feature_columns['LIVER'] = np.where(feature_columns['LIVER'] == 'Yes', 1, 0)
feature_columns['LUNG'] = np.where(feature_columns['LUNG'] == 'Yes', 1, 0)
feature_columns['Regional'] = 0
feature_columns['Distant'] = 0

# create binary columns for clinical summary
# if CLINICAL_SUMMARY contains regional, set Regional to 1
# if CLINICAL_SUMMARY contains distant, set Distant to 1

feature_columns['Regional'] = np.where(
	(feature_columns['CLINICAL_SUMMARY'] == 'Regional,Extension and Nodes') |
	(feature_columns['CLINICAL_SUMMARY'] == 'Regional,Lymph Nodes Only') |
	(feature_columns['CLINICAL_SUMMARY'] == 'Regional,Direct Extension') |
	(feature_columns['CLINICAL_SUMMARY'] == 'Regional To Lymph Nodes') |
	(feature_columns['CLINICAL_SUMMARY'] == 'Regional Both 2 and 3') |
	(feature_columns['CLINICAL_SUMMARY'] == 'Regional by Direct Extension') |
	(feature_columns['CLINICAL_SUMMARY'] == 'Regional Nos'),
	1, 0
)
feature_columns['Distant'] = np.where(
	(feature_columns['CLINICAL_SUMMARY'] == 'Distant') |
	(feature_columns['CLINICAL_SUMMARY'] == 'Distant Metastases/Systemic Disease'),
	1, 0
)

# 'Breast Cancer' 'Non-Small Cell Lung Cancer' 'Colorectal Cancer' 'Pancreatic Cancer' 'Prostate Cancer'
feature_columns['CANCER_TYPE_BREAST'] = np.where(
	feature_columns['CANCER_TYPE'] == 'Breast Cancer', 1, 0
)
feature_columns['CANCER_TYPE_LUNG'] = np.where(
	feature_columns['CANCER_TYPE'] == 'Non-Small Cell Lung Cancer', 1, 0
)
feature_columns['CANCER_TYPE_COLON'] = np.where(
	feature_columns['CANCER_TYPE'] == 'Colorectal Cancer', 1, 0
)
feature_columns['CANCER_TYPE_PANCREAS'] = np.where(
	feature_columns['CANCER_TYPE'] == 'Pancreatic Cancer', 1, 0
)
feature_columns['CANCER_TYPE_PROSTATE'] = np.where(
	feature_columns['CANCER_TYPE'] == 'Prostate Cancer', 1, 0
)

# remove the original CANCER_TYPE column
feature_columns = feature_columns.drop(columns=['CANCER_TYPE'], axis=1)
# drop clinical summary as it is now redundant
feature_columns = feature_columns.drop(columns=['CLINICAL_SUMMARY'], axis=1)

print(f'Number of patients left after removing NaN values: {feature_columns.shape[0]}')
```

```python
# non-continuous columns
for col in feature_columns.columns[1:]:  # Skip the first column (patient ID)
    print(f"Column '{col}': {feature_columns[col].unique()}")
    counts = feature_columns[col].value_counts()
    print(f"Counts for {col}:\n{counts}\n")
```

```python
# concatenate the filtered subtypes with the feature columns based on patient ID
final_df = pd.merge(filtered_subtypes_with_id, feature_columns, on='Patient', how='inner')
final_df
```

```python
# save the final dataframe to a CSV file
final_df.to_csv(f'../data/msk_2024_fe_{threshold}.csv', index=False)
print(f"Final dataframe saved to '../data/msk_2024_fe_{threshold}.csv'")
```
