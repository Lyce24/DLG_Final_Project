```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/msk_2024_mutations_clean.csv')
df
```

```python
# create a new column combining gene, mutationtype, varianttype
# Add Gene Nodes
gene_counts = df['Gene'].value_counts(normalize=True)
genes_to_keep = gene_counts[gene_counts >= 0.001].index
cleaned_df = df[df['Gene'].isin(genes_to_keep)]

genes = cleaned_df["Gene"].unique()
# mutation subtypes
cleaned_df["Mutation_Subtype"] = cleaned_df["Gene"] + "_" + cleaned_df["mutationType"] + "_" + cleaned_df["variantType"]
cleaned_df["Mutation_Subtype_With_Chr"] = cleaned_df["Mutation_Subtype"] + "_" + cleaned_df["chr"].astype(str)
mutation_subtypes = cleaned_df["Mutation_Subtype"].unique()
mutation_subtypes_with_chr = cleaned_df["Mutation_Subtype_With_Chr"].unique()

print("Number of unique genes:", len(genes))
print("Number of unique mutation subtypes:", len(mutation_subtypes))
print("Number of unique mutation subtypes with chr:", len(mutation_subtypes_with_chr))
cleaned_df
```

```python
# save the featured data
cleaned_df.to_csv('../data/msk_2024_mutations_featured.csv', index=False)
```
